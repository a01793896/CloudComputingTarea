```python
# ==========================================
# 1. Instalar librería de Azure Speech
# ==========================================
!pip install azure-cognitiveservices-speech

# ==========================================
# 2. Importar librerías
# ==========================================
import azure.cognitiveservices.speech as speechsdk
from google.colab import files

# ==========================================
# 3. Subir archivo de audio
# ==========================================
print("Sube un archivo de audio WAV")
uploaded = files.upload()

audio_file = list(uploaded.keys())[0]

# ==========================================
# 4. Configuración de Azure Speech
# ==========================================
speech_key = "AJHLWu50haprkHtmZJuNwPxDICbWrOnCG9n6usULtefgs8qgDemrJQQJ99CCACYeBjFXJ3w3AAAYACOG852u"
service_region = "eastus"  # Cambiar por la región de tu recurso

speech_config = speechsdk.SpeechConfig(
    subscription=speech_key,
    region=service_region
)

speech_config.speech_recognition_language = "es-MX"

# ==========================================
# 5. Configurar archivo de audio
# ==========================================
audio_config = speechsdk.audio.AudioConfig(filename=audio_file)

# Crear reconocedor
speech_recognizer = speechsdk.SpeechRecognizer(
    speech_config=speech_config,
    audio_config=audio_config
)

# ==========================================
# 6. Ejecutar reconocimiento
# ==========================================
print("Procesando audio...")

result = speech_recognizer.recognize_once()

# ==========================================
# 7. Mostrar resultados
# ==========================================
if result.reason == speechsdk.ResultReason.RecognizedSpeech:
    print("Texto reconocido:")
    print(result.text)

elif result.reason == speechsdk.ResultReason.NoMatch:
    print("No se reconoció voz.")

elif result.reason == speechsdk.ResultReason.Canceled:
    cancellation_details = result.cancellation_details
    print("Error en el reconocimiento:", cancellation_details.reason)
```
